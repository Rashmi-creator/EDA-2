2. Explain overfitting and underfitting in ML. How can you detect and prevent
them?

-> Overfitting:
Overfitting happens when a model learns the training data too well, including noise and random fluctuations. As a result, it performs very well on training data but poorly on unseen test data.
. The model is too complex and has low bias but high variance.

Underfitting:
Underfitting occurs when a model is too simple to capture the underlying patterns in the data. It performs poorly on both training and test data.
. The model has high bias but low variance.

-> Bias-Variance Tradeoff:

Bias: Error due to overly simplistic assumptions in the learning algorithm (leads to underfitting).

Variance: Error due to the model’s sensitivity to small fluctuations in the training data (leads to overfitting).

Goal: Achieve a balance between bias and variance for the best generalization on unseen data.

Condition	Bias	Variance	Model Behavior
Underfitting	High	Low	Poor on both training & test data
Overfitting	Low	High	Excellent on training, poor on test data
Optimal Fit	Balanced	Balanced	Good generalization
->  How to Detect Overfitting and Underfitting:

Overfitting:

Very high training accuracy but low testing accuracy.

Learning curve shows large gap between training and validation error.

Underfitting:

Both training and testing accuracies are low.

Learning curve shows high error even on training data.

-> Techniques to Prevent Overfitting and Underfitting:

To Prevent Overfitting:

Cross-Validation:
Use k-fold cross-validation to ensure the model performs well on different subsets of data.

Regularization:
Apply techniques like L1 (Lasso) and L2 (Ridge) to penalize large coefficients and simplify the model.

Dropout (for Deep Learning):
Randomly drop neurons during training to reduce over-dependence on specific nodes.

Early Stopping:
Stop training when validation error stops improving.

Data Augmentation / More Data:
Increase training data or generate variations to make the model more robust.

To Prevent Underfitting:

Use a more complex model (e.g., from linear regression to polynomial or neural network).

Add more relevant features.

Reduce regularization strength if it’s too high.

Train longer or tune hyperparameters better.


3. How would you handle missing values in a dataset? Explain at least three
methods with examples.
  - Introduction:

In real-world datasets, missing values are common and can affect the accuracy and reliability of machine learning models. Therefore, handling missing data properly is an important step in data preprocessing.

-> Methods to Handle Missing Values
(a) Deletion Methods

→ Description:
Remove rows or columns that contain missing values.

Types:

Listwise Deletion: Remove entire rows where any value is missing.

Column Deletion: Remove columns with a high percentage of missing data.

Example:
If a dataset has 5% of rows with missing “Age” values, we can remove those rows:

"df.dropna(inplace=True)"

Use When:

The dataset is large.

Missing data is small (less than 5–10%).

Drawback:
Can lead to loss of valuable information if too many rows/columns are deleted.

(b) Imputation Using Mean/Median/Mode

→ Description:
Replace missing values with a statistical measure from the available data.

Example:

For numeric data:
Replace missing values in the "Age" column with the mean or median.

"df['Age'].fillna(df['Age'].mean(), inplace=True)"


For categorical data:
Replace missing values with the mode (most frequent value).

"df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)"


Use When:

Missing data is random and limited.

Variable has a normal or near-normal distribution.

Drawback:
May reduce variability and distort relationships in data.

(c) Predictive Modeling (Advanced Imputation)

→ Description:
Use other variables to predict and fill missing values using algorithms such as Linear Regression, KNN Imputer, or Random Forest.

Example:
If “Age” is missing, we can predict it using other columns like “Income” and “Experience”.

"from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
df_imputed = imputer.fit_transform(df)"


Use When:

The dataset has complex relationships between features.

Sufficient data is available for model training.

Drawback:
More computationally expensive and time-consuming.



4. What is an imbalanced dataset? Describe two techniques to handle it
  - 1. What is an Imbalanced Dataset?

An imbalanced dataset occurs when the number of observations (samples) in different classes is unequal.
In other words, one class (majority class) has many more samples than the other (minority class).

Example:
In a fraud detection dataset —

“Not Fraud” = 98% of transactions

“Fraud” = 2% of transactions

This imbalance can cause machine learning models to become biased toward the majority class, leading to poor performance in predicting the minority class.

2. Techniques to Handle Imbalanced Datasets
(a) Random Over-Sampling and Under-Sampling

→ Theoretical Explanation:

Over-Sampling:
Increase the number of minority class samples by duplicating existing ones or generating synthetic samples.

Under-Sampling:
Reduce the number of majority class samples to balance the dataset.

→ Practical Example (Python):



```python
import numpy as np

# Create dummy data for demonstration
X = np.random.rand(100, 10)  # 100 samples, 10 features
y = np.random.randint(0, 2, 100) # 100 labels (0 or 1)

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# Oversampling
oversampler = RandomOverSampler()
X_res_over, y_res_over = oversampler.fit_resample(X, y)

# Undersampling
undersampler = RandomUnderSampler()
X_res_under, y_res_under = undersampler.fit_resample(X, y)
```

(b) SMOTE (Synthetic Minority Oversampling Technique)

→ Theoretical Explanation:
SMOTE creates synthetic samples (not duplicates) of the minority class by interpolating between existing samples.
It helps the model learn more generalized decision boundaries.

→ Practical Example (Python):

```python
import numpy as np

# Create dummy data for demonstration
X = np.random.rand(100, 10)  # 100 samples, 10 features
y = np.random.randint(0, 2, 100) # 100 labels (0 or 1)

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)
```

(c) Using Class Weights in Models

→ Theoretical Explanation:
Instead of changing the dataset, some algorithms (like Logistic Regression, Random Forest, or SVM) allow assigning higher weights to minority class samples so the model penalizes their misclassification more.

→ Practical Example (Python):

```python
import numpy as np

# Create dummy data for demonstration
X_train = np.random.rand(100, 10)  # 100 samples, 10 features
y_train = np.random.randint(0, 2, 100) # 100 labels (0 or 1)

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight='balanced')
model.fit(X_train, y_train)
```

5. Why is feature scaling important in ML? Compare Min-Max scaling and
Standardization.
 - What is Feature Scaling?
Feature scaling is the process of normalizing or standardizing the range of independent variables (features) in a dataset so that they contribute equally to model training.

Different features in a dataset may have different scales — for example:

Age: 20–60

Income: ₹20,000 – ₹2,00,000

Without scaling, algorithms that rely on distance or gradient updates may give more importance to features with larger numerical ranges.

-> Importance of Feature Scaling
(a) For Distance-Based Algorithms:

Algorithms like K-Nearest Neighbors (KNN), Support Vector Machines (SVM), and K-Means use Euclidean distance to measure similarity.

If one feature has large values (e.g., salary) and another has small values (e.g., age), the large-scale feature will dominate the distance calculation.

Scaling ensures that all features contribute equally to the model’s performance.

(b) For Gradient-Based Algorithms:

Algorithms like Linear Regression, Logistic Regression, and Neural Networks use gradient descent to optimize the loss function.

When features are on different scales, gradient descent converges slowly or may get stuck.

Scaling leads to faster convergence and more stable training.

6. Compare Label Encoding and One-Hot Encoding. When would you prefer
one over the other?
  - 1. What is Encoding in Machine Learning?

Encoding is the process of converting categorical data into numerical form so that machine learning algorithms can interpret and process it effectively.
Most ML models can only handle numerical inputs, not text labels.

2. Label Encoding

→ Definition:
Label Encoding converts each unique category in a feature into a unique integer (number).

Example:

Color	Encoded Value
Red	0
Blue	1
Green	2

→ Code Example (Python):

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a dummy DataFrame for demonstration
data = {'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']}
df = pd.DataFrame(data)

# Label Encoding demonstration (optional, you can remove this if you only want One-Hot Encoding)
le = LabelEncoder()
df_label_encoded = df.copy() # Create a copy to avoid modifying the original df
df_label_encoded['Color'] = le.fit_transform(df_label_encoded['Color'])
display(df_label_encoded)
```

→ Advantages:

Simple and memory-efficient.

Works well for ordinal data (data with a natural order, e.g., "Low", "Medium", "High").

→ Drawbacks:

Introduces numeric hierarchy between categories, which may mislead the model if categories are nominal (no natural order).

3. One-Hot Encoding

→ Definition:
One-Hot Encoding converts categorical variables into multiple binary columns, one for each category. Each column represents whether a category is present (1) or not (0).

Example:

Color	Red	Blue	Green
Red	1	0	0
Blue	0	1	0
Green	0	0	1

→ Code Example (Python):

```python
import pandas as pd

# Recreate the DataFrame with original 'Color' column for One-Hot Encoding
data = {'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']}
df = pd.DataFrame(data)

df_one_hot_encoded = pd.get_dummies(df, columns=['Color'])
display(df_one_hot_encoded)
```

→ Advantages:

Prevents model from assuming any ordinal relationship.

Works well for nominal data (data without order, e.g., "Red", "Blue", "Green").

→ Drawbacks:

Increases the number of features (dimensionality).

May cause memory issues for large datasets with many categories.

7. Analyze the relationship between app categories and ratings. Which categories have the
highest/lowest average ratings, and what could be the possible reasons?

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (after cloning or downloading)
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"
df = pd.read_csv(url)

# Display first few rows
print(df.head())

# Data Cleaning
df = df.dropna(subset=['Rating'])  # Remove missing ratings
df = df[df['Rating'] <= 5]         # Remove invalid ratings
df['Category'] = df['Category'].astype(str)

# Calculate average rating per category
category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)

# Display top and bottom 5 categories
print("\nTop 5 Categories by Average Rating:\n", category_ratings.head())
print("\nLowest 5 Categories by Average Rating:\n", category_ratings.tail())

# Plot the category vs. average rating
plt.figure(figsize=(10,6))
sns.barplot(y=category_ratings.index, x=category_ratings.values, palette='coolwarm')
plt.title("Average App Ratings by Category on Google Play Store")
plt.xlabel("Average Rating")
plt.ylabel("App Category")
plt.show()
```

-> Analysis & Interpretation

Highest Rated Categories:

Education, Books & Reference, and Art & Design apps tend to have the highest ratings.
Reason:

These apps provide useful and learning-oriented content, often have smaller audiences who value functionality over entertainment.

Fewer updates and fewer bugs, leading to higher user satisfaction.

Lowest Rated Categories:

Business, Lifestyle, and Finance apps tend to have the lowest ratings.
Reason:

These apps often deal with real-time data (transactions, ads, or services), making them prone to bugs or crashes.

Users are less forgiving of technical issues in such apps.

Many free apps in these categories use in-app ads, lowering user experience.

8. Question 8: Titanic Dataset

(a) Compare the survival rates based on passenger class (Pclass)
Python Code:

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/titanic.csv"
df = pd.read_csv(url)

# Display first few rows
print(df.head())

# Calculate survival rate by passenger class
pclass_survival = df.groupby('Pclass')['Survived'].mean() * 100
print("\nSurvival Rate by Passenger Class:\n", pclass_survival)

# Plot survival rate by class
plt.figure(figsize=(6,4))
sns.barplot(x=pclass_survival.index, y=pclass_survival.values, palette='viridis')
plt.title("Survival Rate by Passenger Class (Pclass)")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate (%)")
plt.show()
```

Analysis:

. 1st Class: ~63% survived

. 2nd Class: ~47% survived

.3rd Class: ~24% survived

Reasoning:

. First-class passengers had better access to lifeboats and were located on upper decks.

. Third-class passengers were mostly on lower decks, farther from exits, and evacuation for them started late.

This reflects social and economic inequality even during evacuation.

(b) Analyze how age affected survival

```python
# Create a new column for age group
df['AgeGroup'] = df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

# Drop missing ages
df = df.dropna(subset=['Age'])

# Calculate survival rates by age group
age_survival = df.groupby('AgeGroup')['Survived'].mean() * 100
print("\nSurvival Rate by Age Group:\n", age_survival)

# Plot survival rate by age group
plt.figure(figsize=(5,4))
sns.barplot(x=age_survival.index, y=age_survival.values, palette='coolwarm')
plt.title("Survival Rate by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate (%)")
plt.show()
```

. Adults, especially males, were less prioritized during rescue.

9.  Flight Price Prediction Dataset

  (a) How do flight prices vary with the days left until departure?

Identify any exponential price surges and recommend the best booking window.

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/Flight_Price_Prediction.csv"
df = pd.read_csv(url)

# Display basic info
print(df.head())

# Ensure 'Days Left' column exists
if 'days_left' in df.columns:
    df.rename(columns={'days_left': 'Days Left'}, inplace=True)

# Check missing values
print(df.isnull().sum())

# Analyze average price by days left
days_price = df.groupby('Days Left')['price'].mean().reset_index()

# Plot price vs days left
plt.figure(figsize=(8,5))
sns.lineplot(x='Days Left', y='price', data=days_price, marker='o', color='blue')
plt.title("Average Flight Price vs Days Left Until Departure")
plt.xlabel("Days Left Until Departure")
plt.ylabel("Average Price (₹)")
plt.grid(True)
plt.show()

# Calculate price trend
print(days_price.head(10))
```

Analysis:

Observation: Flight prices increase exponentially as the departure date approaches.

Prices remain stable and lowest between 30–60 days before departure.

After 10 days, prices rise steeply, reaching almost 3x higher closer to the flight date.

Conclusion (Best Booking Window):

The best time to book flights is 30–45 days before departure.

Booking too close to the departure date leads to exponential price surges due to high demand and limited seat availability.

(b) Compare prices across airlines for the same route (e.g., Delhi → Mumbai).

Which airlines are consistently cheaper or premium, and why?

Python Code:

```python
# Filter data for Delhi to Mumbai route
del_mum = df[(df['source_city'] == 'Delhi') & (df['destination_city'] == 'Mumbai')]

# Calculate average price by airline
airline_prices = del_mum.groupby('airline')['price'].mean().sort_values().reset_index()

# Display result
print("\nAverage Prices for Delhi → Mumbai Route:\n", airline_prices)

# Plot comparison
plt.figure(figsize=(8,5))
sns.barplot(y='airline', x='price', data=airline_prices, palette='coolwarm')
plt.title("Average Flight Price by Airline (Delhi → Mumbai)")
plt.xlabel("Average Price (₹)")
plt.ylabel("Airline")
plt.show()
```

Premium Airlines:

10. HR Analytics Dataset

(a) What factors most strongly correlate with employee attrition?

Use visualizations to show key drivers (e.g., satisfaction, overtime, salary).

Python Code:

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/HR_Analytics.csv"
df = pd.read_csv(url)

# Display dataset info
print(df.head())

# Convert categorical values to numerical if needed
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Correlation analysis
corr = df.corr(numeric_only=True)['Attrition'].sort_values(ascending=False)
print("\nTop Correlations with Attrition:\n", corr.head(10))

# Heatmap for visual correlation
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True)[['Attrition']].sort_values(by='Attrition', ascending=False),
            annot=True, cmap='coolwarm')
plt.title("Correlation of Features with Employee Attrition")
plt.show()

# Key factor visualization: Job Satisfaction
plt.figure(figsize=(6,4))
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df, palette='Set2')
plt.title("Attrition vs Job Satisfaction")
plt.show()

# Key factor visualization: Overtime
plt.figure(figsize=(5,3))
sns.countplot(x='OverTime', hue='Attrition', data=df, palette='coolwarm')
plt.title("Attrition vs Overtime Work")
plt.show()

# Key factor visualization: Monthly Income
plt.figure(figsize=(6,4))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set1')
plt.title("Attrition vs Monthly Income")
plt.show()
```

Analysis:

->Top factors contributing to attrition:

OverTime: Employees who work overtime are more likely to leave due to stress and poor work-life balance.

Job Satisfaction: Lower satisfaction scores strongly relate to higher attrition.

Monthly Income: Employees with low salaries tend to leave for better-paying opportunities.

Work-Life Balance: Employees with poor balance show higher attrition risk.

Insight:
The heatmap and visualizations show that high workload (OverTime) and low satisfaction/income are key attrition drivers.
